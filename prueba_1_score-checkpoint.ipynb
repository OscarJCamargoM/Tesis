# Base para regresión

**Se empieza a trabajar con una base de datos que se obtuvo haciendo un muestreo aleatorio de las alertas generadas para los escenarios de monitoreo local, entre enero de 2022 y enero de 2023.**

**También, se tiene la variable de respuesta que nos define si la operación se desestimó o si por el contrario pasó a etapa de investigación.**

**Descripción de las variables**

**1- CODESCEN:** Código del escenario de monitoreo por el que se analizó la operación.

**2- NUMCLIEN:** Número de cliente (dato confidencial)

**3- CODIDENT:** Tipo de documento (dato confidencial)

**4- CLAIDENT:** Número de documento (dato confidencial)

**5- TipoPersona:** Tipo de persona. Natural o Jurídica

**6- Edad:** Edad del cliente, ya sea persona natural o jurídica

**7- Antiguedad:** Antigüedad de la vinculación con la entidad bancaria

**8- ActualizacionInf:** Actualización de la información 

**9- S_NrRBA:** Nivel de riesgo cliente según Modelo Gestión del Riesgo

**10- S_NrOfSegment:** Nivel de Riesgo Oficina Gestora según Modelo Gestión del Riesgo

**11- SumOperaciones:** Movimiento crédito ultimos 6 meses

**12- Efectivo:** Cantidad de registros en los que se ha incluido en el informe de operaciones en efectivo en los últimos 3 años

**13- Divisas:** Cantidad de registros en los que se ha incluido en el informe de operaciones en divisas en los últimos 2 años

**14- Pep:** Indica si el cliente es PEP, Relacionado PEP o No PEP

**15- Baep:** Tiene registros en la base BAEP

**16- Alertas:** Cantidad de alertas en la tabla de antecedentes

**17- Investigaciones:** Cantidad de registros de operaciones investigadas en la tabla de antecedentes

**18- Ros:** Cantidad de registros de operaciones sospechosas en la tabla de antecedentes

**19- SumScore:** Score actual

**20- Score:** Decisión de la operación (Investigar o Desestimar)

### Modelo de Regresión Logística

```python
import statsmodels.api as sm
import scipy.stats as stats
#Elegimos la base estandarizada como fuente principal para los modelos
X = standardized_df

# Dividir los datos en conjuntos de entrenamiento y prueba 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Crear el modelo de regresión logística 
logreg = LogisticRegression()

# Entrenar el modelo utilizando los datos de entrenamiento 
logreg.fit(X_train, y_train)

# Realizar predicciones sobre los datos de prueba 
predicciones_rl = logreg.predict(X_test)
```

```python
#Obtener los valores p
coef = logreg.coef_[0]
se = np.sqrt(np.diag(np.linalg.inv(np.dot(X_train.T, X_train))))
z_values = coef/se
p_values = 2 * (1 - stats.norm.cdf(np.abs(z_values)))
```

```python
print(p_values)
```

```python
#Agregar una variable para el intercepto del modelo
Z = sm.add_constant(X)

#Ajustar el modelo de regresión logística
model = sm.Logit(y, Z)
result = model.fit()

#Obtener los valores p
p_values = result.pvalues
print(p_values)
```

```python
# Calcular las métricas del modelo
accuracy_rl = accuracy_score(y_test, predicciones_rl)
recall_rl = recall_score(y_test, predicciones_rl)
precision_rl = precision_score(y_test, predicciones_rl)
auc_rl = roc_auc_score(y_test, predicciones_rl)
F1_score_rl = f1_score(y_test, predicciones_rl)
cm_logreg = confusion_matrix(y_test, predicciones_rl)

# Calcular la métrica GINI
# gini = 2 * auc_rl - 1

# Calcular la métrica KS
# ks = max(tpr-fpr)
```

```python
print('Métricas del modelo de Regresión Logística')
print('--------------------------------------------------------------------')
print('Exactitud del modelo de Regresión Logística: ', accuracy_rl)
print('Recall del modelo de Regresión Logística: ', recall_rl)
print('Precisión del modelo de Regresión Logística: ', precision_rl)
print('AUC del modelo de Regresión Logística: ',auc_rl)
print('F1 Score del modelo de Regresión Logística: ',F1_score_rl)
print('--------------------------------------------------------------------')
```
